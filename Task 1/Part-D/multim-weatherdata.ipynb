```python
import requests
import pandas as pd
import time
```

```python
# OpenWeatherMap API key (replace with your own key)
API_KEY = "ee7fe596751626a39a9c477974574888"

# List of 20 Indian cities with their latitude and longitude
cities = {
    "Delhi": (28.6139, 77.2090),
    "Mumbai": (19.0760, 72.8777),
    "Kolkata": (22.5726, 88.3639),
    "Chennai": (13.0827, 80.2707),
    "Bangalore": (12.9716, 77.5946),
    "Hyderabad": (17.3850, 78.4867),
    "Pune": (18.5204, 73.8567),
    "Ahmedabad": (23.0225, 72.5714),
    "Jaipur": (26.9124, 75.7873),
    "Lucknow": (26.8467, 80.9462),
    "Chandigarh": (30.7333, 76.7794),
    "Bhopal": (23.2599, 77.4126),
    "Patna": (25.5941, 85.1376),
    "Indore": (22.7196, 75.8577),
    "Guwahati": (26.1445, 91.7362),
    "Bhubaneswar": (20.2961, 85.8245),
    "Nagpur": (21.1458, 79.0882),
    "Coimbatore": (11.0168, 76.9558),
    "Vishakhapatnam": (17.6868, 83.2185),
    "Thiruvananthapuram": (8.5241, 76.9366)
}

# Function to get real-time weather data
def get_realtime_weather(city, lat, lon):
    url = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={API_KEY}&units=metric"
    response = requests.get(url)
    data = response.json()
    
    if response.status_code == 200:
        return {
            "City": city,
            "Date": pd.Timestamp.now().strftime("%Y-%m-%d"),
            "Max Temperature (°C)": data["main"]["temp_max"],
            "Min Temperature (°C)": data["main"]["temp_min"],
            "Precipitation (mm)": data.get("rain", {}).get("1h", 0),  # Precipitation in last 1 hour
            "Max Wind Speed (m/s)": data["wind"]["speed"]
        }
    else:
        print(f"Error fetching data for {city}: {data}")
        return None

# Collect data for each city
weather_data = []

for city, (lat, lon) in cities.items():
    weather_info = get_realtime_weather(city, lat, lon)
    if weather_info:
        weather_data.append(weather_info)
    time.sleep(1)  # To avoid rate limits

# Convert to DataFrame
df = pd.DataFrame(weather_data)

# Save to CSV
df.to_csv("real_time_weather_data.csv", index=False)

print("Data collection complete. Saved to real_time_weather_data.csv")
```
